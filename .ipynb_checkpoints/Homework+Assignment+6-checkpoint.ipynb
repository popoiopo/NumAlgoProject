# Exercise 1
Let $P_k$ be the $k$th _Legendre polynomial_ (see paragraph 7.3.4) on $[-1, 1]$.

We know that $P_0(x) = 1$, $P_1(x) = x$, and that $P_k$ satisfy the recurrence relation 

$$k P_k(x) =x (2k-1)  P_{k-1}(x) - (k-1) P_{k-2}(x).$$

We will use [_induction_](https://en.wikipedia.org/wiki/Mathematical_induction) to show that the derivatives of the Legendre polynomials satisfy

$$P_k'(1) = k(k+1)/2.$$

## (a) Base case
Show that $P_0'(1) = 0$ and $P_1'(1) = 1$.

$P_0(x) = 1$ -> $P_0'(x) = 0(0+1)/2 = 0$ 

$P_1(x) = x$ -> $P_1'(x) = 1(1+1)/2 = 1$

# Exercise 2

A bacterial population $P$ grows according to the goemetric progression

$$P_t = rP_{t-1}$$

Where r is the growth rate. The following population counts $P_1 ,\ldots, P_8$ (in billions) are observed:

# (a)
Read chapter 6.6 on Nonlinear Least squares. Use the Gauss-Newton Method to fit the model function $f(t, x_1, x_2) = x_1\!\cdot x_2^t$ to the data. Find estimates for the initial population $P_0=x_1$ and the growth rate $r=x_2$.

```python
import scipy.optimize as so
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import IFrame, display, HTML, Markdown
import warnings
warnings.filterwarnings('ignore')


def prettyfie(ax, x, y, title, legendYN="Yes"):

    # Remove borders
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.grid(alpha=0.25)

    # Remove unnecessary ticks
    ax.tick_params(
        axis="both",
        which="both",
        bottom=False,
        top=False,
        labelbottom=True,
        left=False,
        right=False,
        labelleft=True)

    if legendYN == "Yes":
        # Create legend and grid
        ax.legend(framealpha=1, shadow=True)
        ax.legend()

    # Set labels and title
    ax.set_xlabel(x, fontsize=14)
    ax.set_ylabel(y, fontsize=14)
    ax.set_title(title, fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=12)


def printmd(string):
    display(Markdown(string))
```

```python
data = [0.19, 0.36, 0.69, 1.3, 2.5, 4.7, 8.5, 14]
time = np.linspace(0, 7, 8)

x0 = np.ones(2)


def f(x, t, y):
    return x[0] * x[1]**t - y


ltsq = so.least_squares(f, x0, args=(time, data))


def gen_f(x, t):
    return x[0] * x[1]**t


x = np.linspace(-2, 9, 300)
y = [gen_f(ltsq.x, val) for val in x]

df_ltsq = pd.DataFrame(data={
    "Initial Population": [ltsq.x[0]],
    "Growth rate": [ltsq.x[1]]
})
display(HTML(df_ltsq.to_html(index=False)))

fig, ax = plt.subplots(1, 1, figsize=(8, 5), sharey=True)
ax.plot(x, y, '-', label="Fitted function")
ax.plot(time, data, 'ro', label="Data points")
ax.legend()
prettyfie(ax, "Time", "Population", "Data fit (Gauss-Newton)", legendYN="Yes")
plt.show()

#https://scipy-cookbook.readthedocs.io/items/robust_regression.html
```

```python
lg_data = [np.log(t) for t in data]


def fun(t, a, b):  # this is your 'straight line' y=f(x)
    return a * t + b


a, b = so.curve_fit(fun, time, lg_data)[0]

df_ltsq = pd.DataFrame(data={"X0": [b], "a": [a]})
display(HTML(df_ltsq.to_html(index=False)))
printmd("** Transformed data**")
df_ltsq = pd.DataFrame(data={"Time": time, "Log data": lg_data})
display(HTML(df_ltsq.to_html(index=False)))

t_lin = np.linspace(-1, 9, 300)
y_lin = [fun(t, a, b) for t in t_lin]

fig, ax = plt.subplots(1, 1, figsize=(8, 5), sharey=True)
ax.plot(time, lg_data, 'ro', label="Data points")
plt.plot(t_lin, y_lin, '-', label="Data fit")
ax.legend()
prettyfie(ax, "Time","Log data", "Data fit (Linear least squares)", legendYN="Yes")
plt.show()
```

# (a)
Interpolate Runge's function on $[-1,1]$ by a polynomial $p_k$ of degree $k$ for $k = 5,10,15$. Use equally spaced interpolation points. Draw your interpolating polynomials together with $f$.

# (b)
Repeat (a) to find polynomials $q_k$. This time use the Chebyshev points as interpolation points. They are given by

$$t_i = \cos \left( \frac{(2i-1)\pi}{2k+2} \right),\, \,i = 1,\ldots,k+1.$$

```python
import math


class Chebyshev:
    """
    Chebyshev(a, b, n, func)
    Given a function func, lower and upper limits of the interval [a,b],
    and maximum degree n, this class computes a Chebyshev approximation
    of the function.
    Method eval(x) yields the approximated function value.
    """

    def __init__(self, a, b, n, func):
        self.a = a
        self.b = b
        self.func = func

        bma = 0.5 * (b - a)
        bpa = 0.5 * (b + a)
        f = [
            func(math.cos(math.pi * (k + 0.5) / n) * bma + bpa)
            for k in range(n)
        ]
        fac = 2.0 / n
        self.c = [
            fac * sum([
                f[k] * math.cos(math.pi * j * (k + 0.5) / n) for k in range(n)
            ]) for j in range(n)
        ]

    def eval(self, x):
        a, b = self.a, self.b
        assert (a <= x <= b)
        y = (2.0 * x - a - b) * (1.0 / (b - a))
        y2 = 2.0 * y
        (d, dd) = (self.c[-1], 0)  # Special case first step for efficiency
        for cj in self.c[-2:0:-1]:  # Clenshaw's recurrence
            (d, dd) = (y2 * d - dd + cj, d)
        return y * d - dd + 0.5 * self.c[0]  # Last step is different


def f(x):
    return 1 / (1 + 25 * x**2)


ls = np.linspace(-1, 1, 300)

fig, ax = plt.subplots(1, 1, figsize=(8, 5), sharey=True)
ax.plot(xax, yax, 'o', label="Datapoints")
for val in [5, 10, 15]:
    ch = Chebyshev(-1, 1, val, f)
    listing = [ch.eval(i) for i in ls]
    ax.plot(ls, listing, label=f"Deg: {val}")
    ax.legend()
    prettyfie(
        ax,
        "x",
        "y",
        "Interpolating using the Chebyshev function for polynomial degree 5, 10, 15",
        legendYN="Yes")
plt.show()

# http://excamera.com/sphinx/article-chebyshev.html with much love for these guys
```

# (c)
Verify numerically that the maximum of the difference of $f$ and $q_k$ in the interval $[-1,1]$ decreases with increasing $k$ by trying various values of $k$.

```python
ls = np.linspace(-1, 1, num=300, endpoint=True)
yc = 1 / (1 + 25 * ls**2)
dataCR = {"cheby": {}, "Runge": {}}

for val in [5, 10, 15]:
    ch = Chebyshev(-1, 1, val, f)
    dataCR["cheby"][val] = [ch.eval(i) for i in ls]
    fc = np.polyfit(ls, yc, deg=val)
    pc = np.poly1d(fc)
    dataCR["Runge"][val] = [pc(i) for i in ls]

fig, ax = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
for i, val in enumerate([5, 10, 15]):
    ax[i].plot(ls, dataCR["Runge"][val], label=f"f Deg: {val}")
    ax[i].plot(ls, dataCR["cheby"][val], label=f"qk Deg: {val}")
    ax[i].legend()
    prettyfie(
        ax[i],
        "x",
        "y",
        f"Cheby vs Runge, polynomial degree {val}",
        legendYN="Yes")
plt.show()

print(f"Mean differences between Runge and Chebyshev \
    \nfor degree 5: {abs(np.mean(np.array(dataCR['Runge'][5]) - np.array(dataCR['cheby'][5])))}\
    \nfor degree 10: {abs(np.mean(np.array(dataCR['Runge'][10]) - np.array(dataCR['cheby'][10])))}\
    \nfor degree 15: {abs(np.mean(np.array(dataCR['Runge'][15]) - np.array(dataCR['cheby'][15])))}"
      )
```
